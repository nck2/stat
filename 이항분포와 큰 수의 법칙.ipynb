```python
def mybinom(p=0.2, myn=10,color='black'):
    """횟수 n을 받아서 그래프 반환"""
    m = np.arange(0,myn+1,1)
    mypmf = sp.stats.binom(myn, p=p).pmf(k=m)
    return plt.plot(m, mypmf, color)
```

```python
mybinom(myn=10, color='red')
mybinom(myn=20, color='blue')
mybinom(myn=50, color='gray')
```

```python
mybinom(p=0.5, myn=10, color='red')
mybinom(p=0.5, myn=20, color='blue')
mybinom(p=0.5, myn=50, color='gray')
mybinom(p=0.5, myn=70, color='yellow')
```

### 큰 수의 법칙
어떤 시행에서 사건 $A$가 일어날 확률이 $p$일 때, $n$회의 독립시행에서 사건 $A$가 일어나는 횟수를 $X$라고 하면, 충분히 작은 양수 $h$에 대하여,

$$ \lim_{n \rightarrow \infty} P(|\frac{X}{n} -p|<h) =1$$

#### $X \sim B(n, \frac{1}{6} )$에 대하여 위의 정리를 확인해보자

```python
h = 0.1
p = 1/6
```

```python
n= 10
```

```python
abs(-2)
```

```python
def disx(val, n, h, p):
    """확률변수의 값이 원하는 범위에 들었는지 판단"""
    return abs((val/n) - p)< h
```

```python
sp.stats.binom.pmf(k=1, n=10, p=p)+sp.stats.binom.pmf(k=2, n=10, p=p)
```

```python
def cal(num,p,h):
    result = 0
    for i in range(num+1):
        if disx(i,num,h,p):
            result = result + sp.stats.binom.pmf(k=i, n=num, p=p)
    return result
```

#### $p=1/6 ,~ h=0.1$ 인 상태에서 $n$ 을 증가시키며 값을 확인해보기

```python
cal(10,1/6,0.1)
```

```python
cal(20,1/6,0.1)
```

```python
cal(30,1/6,0.1)
```

```python
cal(50,1/6,0.1)
```
