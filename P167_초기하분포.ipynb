$$ P(X=x) = \frac{~_M C_x ~_{N-M} C_{n-x} }{_N C_n }$$

```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from scipy.stats import hypergeom
import seaborn as sns
sns.set()
```

```python
[N, M, n] = [20, 7, 12]
rv = hypergeom(N, M, n)
x = np.arange(0, n+1)
mypmf = rv.pmf(x)
```

```python
plt.plot(x, mypmf, color='0.5')
```

```python
def myhyper(N,M,n):
    rv = hypergeom(N, M, n)
    prob = M/N
    x = np.arange(0, n+1)
    mypmf = rv.pmf(x)
    return plt.plot(x, mypmf, label=f"p:{prob}")
```

```python
myhyper(20, 7, 12)
```

```python
myhyper(50, 10, 5)
myhyper(50, 25, 5)
myhyper(50, 40, 5)
myhyper(50, 45, 5)
plt.legend()
```

n을 변화

```python
myhyper(50, 10, 20)
myhyper(50, 25, 20)
myhyper(50, 40, 20)
myhyper(50, 45, 20)
myhyper(50, 49, 20)

plt.legend()
```

## 초기하분포와 이항분포 비교

```python
x = np.arange(0, 21)
by = sp.stats.binom(20,0.5).pmf(x)
hpy = hypergeom(60, 30, 20).pmf(x)
plt.plot(x, by, label="binomial")
plt.plot(x, hpy, label="hyper")
# n/N =  1/3
```

```python
x = np.arange(0, 21)
by = sp.stats.binom(20,0.3).pmf(x)
hpy = hypergeom(300, 90, 20).pmf(x)
plt.plot(x, by, label="binomial")
plt.plot(x, hpy, label="hyper")
# n/N =  1/15
```

#### 결론:  n/N 의 값이 충분히 작으면 이항분포와 초기하분포는 비슷하다.
