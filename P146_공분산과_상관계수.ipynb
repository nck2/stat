```python
X,Y =[],[]
for i in range(mytimes):
    dice=np.random.choice(['H','T'],size=3)
    X.append(Counter(dice[:2])['H']) # 2번 던졌을 때 앞면의 수
    Y.append(Counter(dice)['H']) # 3번 던졌을 때 앞면의 수
```

```python
def cov_print(mytimes):
    X,Y =[],[]
    for i in range(mytimes):
        dice=np.random.choice(['H','T'],size=3)
        X.append(Counter(dice[:2])['H']) # 2번 던졌을 때 앞면의 수
        Y.append(Counter(dice)['H']) # 3번 던졌을 때 앞면의 수
    mycov = np.cov(X,Y,ddof=0)[0,1]
    mycoef = np.corrcoef(X,Y)[0,1]
    return f"공분산은 {mycov}이고 상관계수는 {mycoef}이다."
    

    
```

```python
cov_print(1000)
```

```python
X,Y =[],[]
for i in range(mytimes):
    dice=np.random.choice(['H','T'],size=3)
    X.append(Counter(dice[:2])['H']) # 2번 던졌을 때 앞면의 수
    Y.append(Counter(dice)['H']) # 3번 던졌을 때 앞면의 수
```

```python
raw_data={'x':X,'y':Y}
```

```python
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = DataFrame(raw_data)
data
```

```python
df = data.corr()
```

```python
df
```

```python
fig, ax = plt.subplots( figsize=(3,3) )

# 삼각형 마스크를 만든다(위 쪽 삼각형에 True, 아래 삼각형에 False)
mask = np.zeros_like(df, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# 히트맵을 그린다
sns.heatmap(df, 
            cmap = 'RdYlBu_r', 
            annot = True,   # 실제 값을 표시한다
            mask=mask,      # 표시하지 않을 마스크 부분을 지정한다
            linewidths=.5,  # 경계면 실선으로 구분하기
            cbar_kws={"shrink": .5},# 컬러바 크기 절반으로 줄이기
            vmin = 0,vmax = 3   # 컬러바 범위 -1 ~ 1
           )  
plt.show()
```
