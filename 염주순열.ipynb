```python
def next_pe(mylist):
    return [mylist[-1]]+mylist[:-1]
```

```python
next_pe(['a','b','c'])
```

```python
def dis(a,b):
    """
    입력: 두개의 리스트
    출력: 염주순열로 같으면 True 
    """
    temp=a[:]
    for i in range(len(a)):
#         print("temp:",temp,"b",b)
        if temp==b:
            return True
        else:
            temp = next_pe(temp)
    temp2=a[::-1]
    for j in range(len(a)):
        if temp2==b:
            return True
        else:
            temp = next_pe(temp)
    
    
    return False
            
```

```python
dis(['a','b','c',"d"],['d','c','b','a'])
```

```python
ball_perm_list =list(ball_perm)
```

```python
myresult={}
j=0
while ball_perm_list:
    j=j+1
    
    myselect = ball_perm_list.pop()
    myresult[j]=[myselect]
    for i in ball_perm_list:
        if dis(list(i),list(myselect)):
            myresult[j].append(i)
        
    

    
```
